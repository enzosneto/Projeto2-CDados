## Projeto 2 - Ciência dos Dados

```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
```

```python
# Lendo o banco de dados
data_0 = pd.read_csv("data.csv")
```

```python
# Rodando as primeiras 5 linhas do banco de dados
data_0.head()
```

```python
# Filtrando as colunas
data = data_0.loc[:,["song_title", "artist", "tempo", "speechiness", "loudness", "energy", "instrumentalness", "duration_ms", "danceability", "acousticness", "key"]]
```

```python
# Rodando as primeiras linhas depois de filtrar
data.head()
```

```python
# Substituindo os códigos números pelas respectivas notas na coluna "key"
data['key'] = data['key'].astype('category')
data['key'].cat.categories = ['C', 'C#', 'D','D#', 'E','F','F#','G','G#','A','A#','B',]
data.head()
```
